```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Aman", "Saurav", "Karan", "Shubham", "Abhishek",
             "Rohit", "Priya", "Neha", "Arjun", None],
    
    "Department": ["IT", "Sales", "HR", "IT", "Finance",
                   "HR", "IT", "Sales", "Finance", "IT"],
    
    "Role": ["Developer", "Manager", "Analyst", "Developer", "Manager",
             "Analyst", "Developer", "Manager", "Analyst", "Developer"],
    
    "Experience": [2, 5, 3, 4, 6, 2, 1, 7, 5, 0],
    
    "Salary": [60000, 75000, 50000, 65000, 80000,
               52000, 58000, 90000, 72000, 45000],
    
    "Score": [4.5, 4.2, np.nan, 4.8, 4.1,
              3.9, 4.6, 4.3, 4.0, np.nan]
}

df = pd.DataFrame(data)

df
```

```python
# 1️⃣ Average Salary for each Department
df.groupby("Department")["Salary"].mean()

# 2️⃣ Average Performance Score for each Role
df.groupby("Role")["Score"].mean()

# 3️⃣ Non-null Performance Score count for each Department
df.groupby("Department")["Score"].count()

# 4️⃣ Total number of employee records in each Department (including missing values)
df.groupby("Department").size()
```

```python
# Average Salary for each Department
df.groupby("Department")["Salary"].mean()

# Average Performance Score for each Role
df.groupby("Role")["Score"].mean()

# Non-null Performance Score count for each Department
df.groupby("Department")["Score"].count()

# Total number of employee records in each Department
df.groupby("Department").size()
```

```python
# Salary summary (mean, max, min, count) by Department
df.groupby("Department")["Salary"].agg(["mean", "max", "min", "count"])

# Performance Score summary (mean, max, min, count) by Role
df.groupby("Role")["Score"].agg(["mean", "max", "min", "count"])

# Average Salary grouped by Department and Role
df.groupby(["Department", "Role"])["Salary"].mean()

# Average Performance Score grouped by Department and Experience
df.groupby(["Department", "Experience"])["Score"].mean()
```
