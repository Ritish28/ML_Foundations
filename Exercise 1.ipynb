```python
import numpy as np
```

```python
arr=np.array([1500,3000,5000])
```

```python
daily_sales = np.array([[10,15,8,12],[5,7,4,6],[2,3,2,4]])
```

```python
print("Cost Price Shape:", arr.shape)
print("Cost Price Dimension:", arr.ndim)

print("Daily Sales Shape:", daily_sales.shape)
print("Daily Sales Dimension:", daily_sales.ndim)
```

```python
daily_sales_bonus = daily_sales+2
print(daily_sales_bonus)
```

```python
revenue = daily_sales*arr[:, np.newaxis]
print(revenue)
```

```python
tpdrevenue = revenue.sum(axis=0)
print(tpdrevenue)
```

```python
sqrt_values = np.sqrt(daily_sales)
print(sqrt_values)
```
