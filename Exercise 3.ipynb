```python
import numpy as np

electricity = np.array([
    [120, 130, 125, 140, 150, 160, 155],
    [100, 110, 105, 115, 120, 130, 125],
    [90,  95,  100, 98,  105, 110, 108],
    [140, 145, 150, 155, 160, 165, 170]
])
```

Part A

```python
electricity = np.random.randint(120, 651, size=(4, 7))
print(electricity)
```

```python
print("Shape:", electricity.shape)
print("Number of dimensions:", electricity.ndim)
print("Total number of elements:", electricity.size)
print("Data type:", electricity.dtype)
```

Part B

```python
avg_per_hostel = np.mean(electricity, axis=1)
print("Average electricity per hostel:", avg_per_hostel)
```

```python
max_per_hostel = np.max(electricity, axis=1)
print("Maximum electricity per hostel:", max_per_hostel)
```

```python
highest_avg_hostel = np.argmax(avg_per_hostel)
print("Hostel with highest average usage:", highest_avg_hostel)
```

```python
print("Hostel with highest average usage: Hostel", highest_avg_hostel + 1)
```

Part B2

```python
total_per_day = np.sum(electricity, axis=0)
print("Total electricity per day:", total_per_day)
```

```python
highest_day = np.argmax(total_per_day)
print("Day with highest total consumption: Day", highest_day + 1)
```

Part C

```python
import numpy as np

usage = np.array([
    [500, 600, 450, 700],
    [480, 520, 610, 580],
    [530, 490, 560, 620]
])
```

```python
penalized_usage = np.where(usage > 550, usage * 1.20, usage)

print("Penalized Usage:")
print(penalized_usage)
```

Part D

```python
avg = np.mean(usage, axis=1, keepdims=True)
```

```python
comparison = usage > avg
```

```python
print("Averages:\n", avg)
print("Comparison:\n", comparison)
```
