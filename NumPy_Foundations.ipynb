Where in NumPy

```python
arr=np.array([10,-5,20,-3,15])
```

```python
result = np.where(arr<0,0,arr)
print(result)
```

```python
arr=np.array([1,2,3,4,5])
```

```python
result = np.where(arr%2 != 0, -1, arr)
print(result)
```

```python
marks = np.array([45,78,32,90,55])
```

```python
status = np.where(marks>=50,1,0)
print(status)
```

```python
data = np.array([10,20,30,40,50])
```

```python
mean_val = np.mean(data)
result = np.where(data > mean_val , mean_val , data) #1 parameter is condn , 2nd is true value , 3rd false value
print(result)
```

Applying (Where) condition in 2d array

```python
mat = np.array([[10,60,30] , [80,20,30]])
```

```python
result = np.where(mat >= 50, 1, 0)
print(result)
```
