Part C

```python
marks = np.array([78, 45, 60, 30, 90])
```

```python
status = np.where(marks >= 50, 1, 0)
print(status)
```

```python
cleaned_marks = np.where(marks < 40, 0, marks)
print(cleaned_marks)
```

Part D

```python
avg = np.array([92, 76, 63, 48, 85])
```

```python
score = np.where(avg >= 85, 10,
         np.where(avg >= 70, 8,
         np.where(avg >= 50, 6, 0)))

print(score)
```

```python
grade = np.where(avg >= 85, 'A',
         np.where(avg >= 70, 'B',
         np.where(avg >= 50, 'C', 'No Grade')))

print(grade)
```
