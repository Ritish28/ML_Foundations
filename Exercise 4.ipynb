DataFrame in pandas

```python
data={
    "Name":["Aman","krish","karan"],
    "Age":[22,23,21],
    "Marks":[85,90,88]
}
```

```python
print(data)
```

```python
type(data)
```

```python
marks=pd.DataFrame(data)
```

```python
print(marks) #Pandas automatically adds a default index: this is called a RangeIndex
```

```python
# we can also add custom indewx in pandas dataframe
marks=pd.DataFrame(
    data,
    index=["Student1","Student2","Student3"]
)
```

```python
print(marks)
```

```python
marks.set_index("Name",inplace=True) #if we are setting inplace=True it means we are changing the original object itself.
```

```python
print(marks)
```

Handling Missing values

```python
# updating the dataset with null values
df = pd.DataFrame({
    "Name":["Aman","saurav","Karan","Shubham",None,"sameer"],
    "Age":[22,23,None,22,25,26],
    "Marks":[85,None,88,90,56,None]
})
df
```

```python
df.isnull()#checking the missing values
```

```python
df.isnull().sum() #counting the missing values
```

```python
df.dropna()
```
